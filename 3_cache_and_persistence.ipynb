Read the data

```python
path = 'dbfs:/FileStore/BigMart_Sales.csv'
```

```python
df = spark.read.format('csv').\
                option('header', True).\
                option('inferschema', True).\
                load(path)
```

Read the same df with a filter

```python
from pyspark.sql.functions import *
```

```python
df_1 = df.filter(col('Outlet_Location_Type') == 'Tier 1')
```

```python
df_2 = df.filter(col('Outlet_Location_Type') == 'Tier 2')
```

```python
display(df_2)
```

For calculating df2, it recalculates the df by scanning the csv as seen in the DAG.

**Cache df**

```python
df.cache()
```

```python
# Calculate df_2 again and check the DAG
df_2 = df.filter(col('Outlet_Location_Type') == 'Tier 2')
```

```python
display(df_2)
```

This time inMemoryTable scan happened instead of reading the csv again

```python

df.unpersist()
```

**Store in memory only**

```python
from pyspark.storagelevel import StorageLevel
df.persist(StorageLevel.MEMORY_ONLY)
```
